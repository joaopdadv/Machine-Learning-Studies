```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class PhysicalModel():
    def __init__(self, a=9.81):
        self.a = a

    def forward(self, x):
        return self.a * x
```

```python
trainX = np.linspace(0, 100, 1000)
model = PhysicalModel()
testY = model.forward(trainX)
```

```python
trainX
```

```python
testY
```

```python
class SimpleModel():
    def __init__(self, a=0.):
        self.a = a

    def forward(self, x):
        return self.a * x
    
    def __call__(self, x):
        return self.forward(x)
```

```python

all_a = np.linspace(0, 15, 1000)
all_errors = []
for a in all_a:
    ML_model = SimpleModel(a=a)
    prediction = ML_model(trainX)
    error = np.mean((prediction - testY)**2)
    all_errors.append(error)
```

```python
plt.plot(all_a, all_errors)
```
